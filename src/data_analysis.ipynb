```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_excel('/workspaces/StackBot/src/6.4210_Data.xlsx')
```

```python
df.head()
```

```python
trans_error_diffik = df["DiffIK"]
rot_error_diffik = df["Unnamed: 2"]
trans_error_rrt = df["RRT"]
rot_error_rrt = df["Unnamed: 4"]
```

```python
trans_error_diffik = np.array(pd.to_numeric(trans_error_diffik, errors="coerce").dropna().tolist())
rot_error_diffik = np.array(pd.to_numeric(rot_error_diffik, errors="coerce").dropna().tolist())
trans_error_rrt = np.array(pd.to_numeric(trans_error_rrt, errors="coerce").dropna().tolist())
rot_error_rrt = np.array(pd.to_numeric(rot_error_rrt, errors="coerce").dropna().tolist())
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 4))

axes[0].hist(trans_error_diffik * 100, bins=5, density=True, edgecolor='black', color='skyblue', label='Diff IK')
axes[0].set_title("Differential IK")
axes[0].set_xlabel("Translation Error (mm)")
axes[0].set_ylabel("Probability")

axes[1].hist(trans_error_rrt * 100, bins=5, density=True, edgecolor='black', color='skyblue', label='RRT')
axes[1].set_title("RRT-Connect")
axes[1].set_xlabel("Translation Error (mm)")
axes[1].set_ylabel("Probability")

axes[0].grid(True, linestyle='--', alpha=0.5)
axes[1].grid(True, linestyle='--', alpha=0.5)

axes[0].set_xlim(0, 60)
axes[1].set_xlim(0, 60)

axes[0].set_ylim(0, 0.05)
axes[1].set_ylim(0, 0.05) 

axes[2].hist(rot_error_diffik * 100, bins=5, density=True, edgecolor='black', color='skyblue', label='Diff IK')
axes[2].set_title("Differential IK")
axes[2].set_xlabel("Rotation Error (radians)")
axes[2].set_ylabel("Probability")

axes[3].hist(rot_error_rrt * 100, bins=5, density=True, edgecolor='black', color='skyblue', label='RRT')
axes[3].set_title("RRT-Connect")
axes[3].set_xlabel("Rotation Error (radians)")
axes[3].set_ylabel("Probability")

axes[2].grid(True, linestyle='--', alpha=0.5)
axes[3].grid(True, linestyle='--', alpha=0.5)

axes[2].set_xlim(0, 60)
axes[3].set_xlim(0, 60)

axes[2].set_ylim(0, 0.05)
axes[3].set_ylim(0, 0.05) 


plt.show()
```

```python
plt.hist(trans_error_diffik * 1000, bins=5, density=True, edgecolor='black', color='skyblue', label='Diff IK')
plt.title("Translation Error")
plt.xlabel("Error (mm)")
plt.ylabel("Probability Density")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
